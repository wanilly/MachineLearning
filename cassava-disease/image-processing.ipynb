```python
img = cv2.imread('/content/extra-image-12811.jpg')
mask = np.zeros(img.shape[:2],np.uint8)
```

```python
bgdModel = np.zeros((1,65),np.float64)
fgdModel = np.zeros((1,65),np.float64)
rect = (50,50,450,290)
cv2.grabCut(img,mask,rect,bgdModel,fgdModel,5,cv2.GC_INIT_WITH_RECT)
mask2 = np.where((mask==2)|(mask==0),0,1).astype('uint8')
img = img*mask2[:,:,np.newaxis]
```

```python
image = cv2.imread('/content/extra-image-12811.jpg', cv2.IMREAD_GRAYSCALE)

# 이미지의 모든 행과 열의 절반만 선택
image_cropped = image[300:700,:]

plt.imshow(image_cropped, cmap='gray')
plt.show()
```

```python

# 커널 생성(대상이 있는 픽셀을 강조)
kernel = np.array([[0, -1, 0],
                   [-1, 5, -1],
                   [0, -1, 0]])

# 커널 적용 /
image_sharp = cv2.filter2D(image_cropped, -1, kernel)

fig, ax = plt.subplots(1,2, figsize=(10,5))
ax[0].imshow(image, cmap='gray')
ax[0].set_title('Original Image')
ax[1].imshow(image_sharp, cmap='gray')
ax[1].set_title('Sharp Image')
```

```python
categoryList = [train_cbb, train_cbsd, train_cgm, train_cmd, train_healthy, test, extraimages]
```

```python
# 커널 생성(대상이 있는 픽셀을 강조) 
kernel = np.array([[0, -1, 0],
                   [-1, 5, -1],
                   [0, -1, 0]])
```

```python
# define function
def contour():
  for cate in categoryList:
    for imgPath in cate:
      img = cv2.imread(imgPath) # original image
      imgG = cv2.imread(imgPath, cv2.COLOR_BGR2GRAY)
      imgK = cv2.filter2D(imgG, -1, kernel) # kernel application
      cv2.imwrite('/content/drive/MyDrive/project/MachineLearning/dataset/sample_data/*.jpg', imgK)
```

```python

if __name__ == '__main__':
    contour() # function calling
```
